# Coffee Shop â€” Income Statement

A three-year income statement for a single coffee shop location.

The model is defined in `model.py` so it can be imported here, run as a script, or
served via the Flask explorer app (`explorer`).

Key features demonstrated:
- `Assumption` scalars for COGS rates referenced inside formulas
- Tags (`revenue`, `cogs`, `opex`) driving tag-based subtotals
- `FormulaLine` margin calculations
- `from_template` custom table layout
- Charts

```python
import sys
from pathlib import Path

# Works whether the kernel CWD is the notebook's directory or the project root
_here = Path('.').resolve()
_model_dir = _here if (_here / 'model.py').exists() else _here / 'examples' / 'coffee_shop'
sys.path.insert(0, str(_model_dir))

from model import model
```

## Income statement â€” full table

```python
from pyproforma.tables import BlankRow, HeaderRow, ItemRow, LabelRow, TagTotalRow

table = model.tables.from_template([
    HeaderRow(),
    LabelRow(label="Revenue"),
    ItemRow(name="coffee_sales"),
    ItemRow(name="food_sales"),
    ItemRow(name="merchandise_sales", bottom_border="single"),
    TagTotalRow(tag="revenue", label="Total Revenue", bold=True),
    BlankRow(),
    LabelRow(label="Cost of Goods Sold"),
    ItemRow(name="coffee_cogs"),
    ItemRow(name="food_cogs", bottom_border="single"),
    TagTotalRow(tag="cogs", label="Total COGS", bold=True),
    BlankRow(),
    ItemRow(name="gross_profit", bold=True),
    ItemRow(name="gross_margin"),
    BlankRow(),
    LabelRow(label="Operating Expenses"),
    ItemRow(name="rent"),
    ItemRow(name="labor"),
    ItemRow(name="utilities"),
    ItemRow(name="marketing", bottom_border="single"),
    TagTotalRow(tag="opex", label="Total OpEx", bold=True),
    BlankRow(),
    ItemRow(name="operating_profit", bold=True),
    ItemRow(name="operating_margin"),
])
table.show()
```

## Revenue breakdown

```python
model.tag["revenue"].table(include_label=True).show()
```

## Operating profit â€” period-over-period analysis

```python
model.tables.line_item(
    "operating_profit",
    include_percent_change=True,
    include_cumulative_change=True,
).show()
```

## Charts

```python
# Revenue by stream
model.charts.line_items(
    ["coffee_sales", "food_sales", "merchandise_sales"],
    chart_type="bar",
    title="Revenue by Stream",
).show()
```

```python
# Profit waterfall â€” revenue, gross profit, operating profit
model.charts.line_items(
    ["total_revenue", "gross_profit", "operating_profit"],
    chart_type="line",
    title="Revenue to Profit",
).show()
```

```python
# Margin trend
model.charts.line_items(
    ["gross_margin", "operating_margin"],
    chart_type="line",
    title="Margin Trends",
).show()
```

```python
# OpEx stacked
model.charts.line_items(
    ["rent", "labor", "utilities", "marketing"],
    chart_type="stacked_bar",
    title="Operating Expense Stack",
).show()
```
